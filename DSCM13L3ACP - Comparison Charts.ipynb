```python
labels = ['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','FUELCONSUMPTION_COMB_MPG','CO2EMISSIONS']
```

```python
sns.countplot(data['FUELTYPE'], palette='winter')
```

```python
sns.countplot(data['VEHICLECLASS'])
plt.xticks(rotation=90)
```

```python
df_grouped = data.groupby('FUELTYPE').mean()
df_grouped = df_grouped.reset_index()
```

```python
plt.bar(df_grouped['FUELTYPE'],df_grouped['CO2EMISSIONS'], color='purple')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions')
plt.show()
```

```python
# numerical x
x = np.arange(0, len(df_grouped.index))
# plot bars
plt.bar(x, df_grouped['FUELCONSUMPTION_HWY'], color = '#8390FA')
plt.bar(x, df_grouped['FUELCONSUMPTION_CITY'], bottom=df_grouped['FUELCONSUMPTION_HWY'], color = '#1D2F6F')
plt.bar(x, df_grouped['FUELCONSUMPTION_COMB'], bottom=df_grouped['FUELCONSUMPTION_COMB'], color = '#6EAF46')
plt.ylabel('Fuel Consumption')
plt.xticks(x, ['D','E','X','Z'])
plt.legend(['City','Highway','Combined'])
plt.title('Average Fuel Consumption for Different Vehicle Type')
plt.show()
```

```python
df1_grouped = data.groupby('VEHICLECLASS').mean()
df1_grouped = df1_grouped.reset_index()
```

```python
plt.bar(df1_grouped['VEHICLECLASS'],df1_grouped['CO2EMISSIONS'], color='teal')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions')
plt.xticks(rotation=90)
plt.show()
```
